```python
import random
import unidecode

def randomly_remove_accents(text, probability=0.5):
    new_text = ""
    for char in text:
        if random.random() < probability:
            new_char = unidecode.unidecode(char)
        else:
            new_char = char
        new_text += new_char
    return new_text
```

```python
# Example usage:
text = "Lâm Hoàng Khang"
new_text = randomly_remove_accents(text)
print(new_text)
```

```python
data['AccentlessSentences'] = data['Sentences'].apply(lambda x: randomly_remove_accents(str(x)))
```
